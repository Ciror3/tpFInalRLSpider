```python
env = SpiderEnv(render_mode=None)

model = PPO(
    "MlpPolicy",
    env,
    learning_rate=0.000355207825977822,
    n_steps=256,                      
    batch_size=8,                     
    n_epochs=20,
    gamma=0.9791381032304535,         
    gae_lambda=0.9975914694455162,    
    clip_range=0.1,
    ent_coef=1.440070886288556e-06,
    vf_coef=0.5,
    max_grad_norm=0.7456096284142648,
    policy_kwargs=dict(
        net_arch=[dict(pi=[64, 64], vf=[64, 64])],  
        activation_fn=torch.nn.Tanh,                
    ),
    verbose=0,
    tensorboard_log="logs_spider",
)

model.learn(total_timesteps=2_000_000)
model.save("ppo_spider")
```
